```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0,6])
ypoints = np.array([1,8])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,x,y2)
plt.show()
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,'g')
plt.plot(x,y2,'o')
plt.show()
```

```python
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,'o')
plt.plot(x,y2,marker='*')
plt.show()
```

```python
x = np.linspace(0,10,20)
y = x**2
plt.plot(x,y,'m-.',marker='o',ms=9)
plt.show()
```

```python
x = np.linspace(0,10,20)
y = x**2
plt.plot(x,y,'m-.',marker='o',ms=9)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("funcion cuadrada", fontdict = font1)
plt.xlabel("x", fontdict = font2)
plt.ylabel("y", fontdict = font2)

plt.show()
```

```python
#subplots
x=np.linspace(0,20,100)
y= x**2

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("funcion cuadrada", fontdict = font1)
plt.xlabel("x", fontdict = font2)
plt.ylabel("y", fontdict = font2)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()

y2=np.sin(x)
plt.title("funcion seno", fontdict = font1)
plt.xlabel("x", fontdict = font2)
plt.ylabel("y", fontdict = font2)
plt.subplot(1,2,1)
plt.plot(x,y2)
plt.show()
```
