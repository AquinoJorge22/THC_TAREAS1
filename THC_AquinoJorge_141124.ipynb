```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
def get_data(N=20):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1
  for i in range(N):
    y[i]=2*x[i]+3+(eps*(-1)**i)
  return x,y
```

```python
x,y=get_data(40)
def plot_data(x,y,params):
  z=params[0]+params[1]*x

  plt.figure()
  plt.plot(x,y,'r*',label='Samples')
  plt.plot(x,z,'-k',label='Model')
  plt.grid(lw=2)
  plt.title('Linear Regression')
  plt.xlabel('x Data')
  plt.ylabel('y Data')
  plt.legend()
  plt.show()
plot_data(x,y,params)
```

```python
def linear_regression(N=20):
  x,y=get_data(N)
  A=np.vstack([np.ones(len(x)),x]).T
  AtA=A.T@A
  Aty=A.T@y
  params=LA.solve(AtA,Aty)
  plot_data(x,y,params)
  print(params)
  return params
```

```python
def get_data(N):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1
  for i in range(N):
    y[i]=3*x[i]**2+5*x[i]-2+5*(eps*(-1)**i)
  return x,y
```

```python
def plot_data(x,y,params):
  z=params[0]+params[1]*x+params[2]*x**2
  plt.figure()
  plt.plot(x,y,'r*',label='Samples')
  plt.plot(x,z,'-k',label='Model')
  plt.grid(lw=2)
  plt.title('Cuadratic Regression')
  plt.xlabel('x Data')
  plt.ylabel('y Data')
  plt.legend()
  plt.show()
```

```python
def cuadratic_regression(N=40):
  x,y=get_data(N)
  A=np.vstack([np.ones(len(x)),x,x**2]).T
  AtA=A.T@A
  Aty=A.T@y
  params=LA.solve(AtA,Aty)
  plot_data(x,y,params)
  #print(params)
  OriData=np.array([-2.,5.,3.])
  Error=np.round(LA.norm(params-OriData),3)
  print('Error: ', Error*100, '%')
  return params
```

```python
cuadratic_regression(60)
```
